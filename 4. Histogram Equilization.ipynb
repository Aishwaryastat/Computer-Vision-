```python
import numpy as np 
import cv2
import matplotlib.pyplot as plt 

image = cv2.imread(r"C:\Users\pcc\Desktop\DSC_3073.JPG",0)
image= cv2.resize(image, (500,500))

equ= cv2.equalizeHist(image)

res = np.hstack((image,equ))

plt.imshow(res)
plt.show()
```

# CLAHE (Control limited Adaptive Histogram Equilization) 

```python
image = cv2.imread(r"C:\Users\pcc\Desktop\DSC_3073.JPG")
image= cv2.resize(image, (500,500))
image_bw = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

clahe= cv2.createCLAHE(clipLimit=3)
final_img=clahe.apply(image_bw) +10

_,ordinary_img = cv2.threshold(image_bw,150,255,cv2.THRESH_BINARY)

plt.imshow(ordinary_img)
plt.show()
plt.imshow(final_img)
plt.show()
```
