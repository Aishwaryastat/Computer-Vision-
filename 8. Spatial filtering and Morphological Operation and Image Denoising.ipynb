Erosion and dilation are basic operations in image processing, particularly in morphological image processing, which involves manipulating the shapes and structures within an image. These operations are often used for tasks such as noise reduction, object detection, and feature extraction. Both erosion and dilation involve using a structuring element, which is a small binary image or a kernel, to modify the pixel values of the input image.

Here's what erosion and dilation do to an image:

1. **Erosion**:
   Erosion is a morphological operation that shrinks or erodes the boundaries of the foreground (object) in a binary image. It works by sliding the structuring element over the image and checking if all the pixels under the structuring element are non-zero (i.e., part of the object). If all pixels match, the center pixel of the structuring element is set to 1 in the output image; otherwise, it is set to 0.

   Erosion is useful for tasks like reducing noise, separating connected objects, and shrinking object boundaries. It can make objects smaller and remove small noise regions.

2. **Dilation**:
   Dilation is another morphological operation that expands or dilates the boundaries of the foreground in a binary image. Like erosion, it involves sliding the structuring element over the image. If at least one pixel under the structuring element is non-zero, the center pixel of the structuring element is set to 1 in the output image; otherwise, it remains 0.

   Dilation is used to enlarge the size of objects, close small gaps between objects, and join nearby objects. It's often used in conjunction with erosion to perform tasks like object detection and image segmentation.

In summary:
- **Erosion** tends to shrink the object and remove small details or noise.
- **Dilation** tends to expand the object and fill in small gaps or holes.

These operations are building blocks for more complex morphological operations and can be combined in various ways to achieve specific results in image processing tasks.

# Erosion

````python
import cv2
import matplotlib.pyplot as plt
import numpy as np 
````

````python
image=cv2.imread(r"C:\Users\pcc\Desktop\img.jpg",0)

binary_image= cv2.threshold(image,0,255,cv2.THRESH_BINARY+cv2.THRESH_OTSU)[1]

kernel = np.ones((5,5),np.uint8)

invert = cv2.bitwise_not(binary_image)

erosion= cv2.erode(invert,kernel,iterations=1)

plt.imshow(image,cmap="gray")
plt.show()
plt.imshow(erosion,cmap="gray")
plt.show()
````

# Dilation

````python
binary_image=cv2.threshold(image,0,255,cv2.THRESH_BINARY+cv2.THRESH_OTSU)[1]
kernel=np.zeros((5,5),np.uint8)
invert=cv2.bitwise_not(binary_image)
dilation=cv2.dilate(invert,kernel,iterations=10)
plt.imshow(image,cmap="gray")
plt.show()
plt.imshow(dilation,cmap="gray")
````

# Opening

 erosion, Dilation = opening

````python
binary_image=cv2.threshold(image,0,255,cv2.THRESH_BINARY+cv2.THRESH_OTSU)[1]
kernel=np.zeros((5,5),np.uint8)
invert=cv2.bitwise_not(binary_image)
opening=cv2.morphologyEx(invert,cv2.MORPH_OPEN,kernel,iterations=10)
plt.imshow(image,cmap="gray")
plt.show()
plt.imshow(opening,cmap="gray")
````

# Closing 

 dilation, erosion = closing

````python
binary_image=cv2.threshold(image,0,255,cv2.THRESH_BINARY+cv2.THRESH_OTSU)[1]
kernel=np.zeros((5,5),np.uint8)
invert=cv2.bitwise_not(binary_image)
closing=cv2.morphologyEx(invert,cv2.MORPH_CLOSE,kernel,iterations=10)
plt.imshow(image,cmap="gray")
plt.show()
plt.imshow(closing,cmap="gray")
````

# Total Variation Denoising 

Total Variation (TV) denoising is a popular technique in image processing for reducing noise while preserving edges and features in an image. It's particularly effective for denoising images that have a high level of noise while maintaining sharp edges and fine details.

The concept behind Total Variation denoising is based on the principle that images with less variation (smoother regions) are likely to have less noise, while abrupt changes (edges and details) in an image correspond to high variation. The method tries to minimize the total variation of the image while keeping the original data as close as possible.

Mathematically, the Total Variation of an image can be defined as the integral of the gradient magnitude:

```
TV(I) = ∫ ||∇I|| dA
```

where:
- `TV(I)` is the Total Variation of the image `I`.
- `∇I` is the gradient of the image.
- `dA` represents the area element.

The goal of Total Variation denoising is to find an image that minimizes the Total Variation while being close to the noisy input image. This can be formulated as an optimization problem:

```
argmin ||I - I0||^2 + λ * TV(I)
```

where:
- `I` is the denoised image.
- `I0` is the noisy input image.
- `λ` is a parameter that balances between the data fidelity term (`||I - I0||^2`) and the Total Variation term (`TV(I)`).

Solving this optimization problem involves iterative methods that update the denoised image to minimize the Total Variation while staying close to the noisy image.

Total Variation denoising is particularly useful for preserving edges and fine details in images, making it suitable for medical imaging, photography, and other applications where preserving structural information is crucial. However, it's important to note that Total Variation denoising can sometimes lead to "staircase" artifacts along edges, which can be mitigated by using more advanced denoising techniques or variations of the Total Variation approach.

## VisuShrink

````python
from skimage.restoration import (denoise_wavelet, estimate_sigma)
from skimage.util import random_noise
from skimage import img_as_float

image=cv2.imread("C:/Users/pcc/Desktop/dog.jpg")
image=cv2.cvtColor(image,cv2.COLOR_BGR2RGB)
image=img_as_float(image)
sigma=0.15

noisy=random_noise(image,var=sigma**2)

sigma_est = estimate_sigma(noisy,average_sigmas=True, multichannel=True)

print(sigma_est)

im_visushrink=denoise_wavelet(noisy,convert2ycbcr=True,sigma=sigma_est,
                              method="VisuShrink",mode="soft",multichannel = True)

fig, ax = plt.subplots(nrows=1, ncols=3, figsize=(8, 5),
                       sharex=True, sharey=True)

ax[0].imshow(image)
ax[0].set_title("Original")

ax[1].imshow(noisy)
ax[1].set_title("Noisy")

ax[2].imshow(im_visushrink)
ax[2].set_title("VisuShrink")

plt.show()
````

# Non Local Mean

````python
import numpy as np
import matplotlib.pyplot as plt
import cv2
from skimage import data, img_as_float
from skimage.restoration import denoise_nl_means, estimate_sigma
from skimage.metrics import peak_signal_noise_ratio
from skimage.util import random_noise
image=cv2.imread("C:/Users/pcc/Desktop/dog.jpg")
image=cv2.cvtColor(image,cv2.COLOR_BGR2RGB)

sigma = 0.08
noisy = random_noise(image, var=sigma**2)

# estimate the noise standard deviation from the noisy image
sigma_est = np.mean(estimate_sigma(noisy,multichannel=True))
print(f'estimated noise standard deviation = {sigma_est}')

patch_kw = dict(patch_size=5,      # 5x5 patches
                patch_distance=6  # 13x13 search area
                )

# slow algorithm
denoise = denoise_nl_means(noisy, h=1.15 * sigma_est, fast_mode=False,
                           **patch_kw)

# slow algorithm, sigma provided
denoise2 = denoise_nl_means(noisy, h=0.8 * sigma_est, sigma=sigma_est,
                            fast_mode=False, **patch_kw)

# fast algorithm
denoise_fast = denoise_nl_means(noisy, h=0.8 * sigma_est, fast_mode=True,
                                **patch_kw)

# fast algorithm, sigma provided
denoise2_fast = denoise_nl_means(noisy, h=0.6 * sigma_est, sigma=sigma_est,
                                 fast_mode=True, **patch_kw)

fig, ax = plt.subplots(nrows=2, ncols=3, figsize=(8, 6),
                       sharex=True, sharey=True)

ax[0, 0].imshow(noisy)
ax[0, 0].axis('off')
ax[0, 0].set_title('noisy')
ax[0, 1].imshow(denoise)
ax[0, 1].axis('off')
ax[0, 1].set_title('non-local means\n(slow)')
ax[0, 2].imshow(denoise2)
ax[0, 2].axis('off')
ax[0, 2].set_title('non-local means\n(slow, using $\\sigma_{est}$)')
ax[1, 0].imshow(image)
ax[1, 0].axis('off')
ax[1, 0].set_title('original\n(noise free)')
ax[1, 1].imshow(denoise_fast)
ax[1, 1].axis('off')
ax[1, 1].set_title('non-local means\n(fast)')
ax[1, 2].imshow(denoise2_fast)
ax[1, 2].axis('off')
ax[1, 2].set_title('non-local means\n(fast, using $\\sigma_{est}$)')

fig.tight_layout()

# print PSNR metric for each case
psnr_noisy = peak_signal_noise_ratio(image, noisy)
psnr = peak_signal_noise_ratio(image, denoise)
psnr2 = peak_signal_noise_ratio(image, denoise2)
psnr_fast = peak_signal_noise_ratio(image, denoise_fast)
psnr2_fast = peak_signal_noise_ratio(image, denoise2_fast)

print(f'PSNR (noisy) = {psnr_noisy:0.2f}')
print(f'PSNR (slow) = {psnr:0.2f}')
print(f'PSNR (slow, using sigma) = {psnr2:0.2f}')
print(f'PSNR (fast) = {psnr_fast:0.2f}')
print(f'PSNR (fast, using sigma) = {psnr2_fast:0.2f}')

plt.show()
````
