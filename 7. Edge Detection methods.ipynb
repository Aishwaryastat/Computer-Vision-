```python
import cv2
import matplotlib.pyplot as plt
import PIL
import numpy as np 
```

# Sobel edge detection

```python
image= cv2.imread(r"C:\Users\pcc\Desktop\colorful.jpg",0)

blur=cv2.GaussianBlur(image,(3,3),0)

# Calculate gradient in x and y directions using Sobel operators
gradient_x=cv2.Sobel(blur,cv2.CV_64F,1,0,ksize=3)
gradient_y=cv2.Sobel(blur,cv2.CV_64F,0,1,ksize=3)

# Calculate the magnitude of gradients
magnitude= np.sqrt(gradient_x**2+gradient_y**2)
# Normalize the magnitude to the range [0, 255]
magnitude = np.uint8(magnitude)

# Apply thresholding to get edges
threshold=50
edges= cv2.threshold(magnitude,threshold,255,cv2.THRESH_BINARY)[1]

plt.imshow(image)
plt.show()

plt.imshow(edges)
plt.show()
```

# Robert Edge Detection

```python
# Apply Roberts edge detection
edges = cv2.filter2D(image, cv2.CV_64F, np.array([[1, 0], [0, -1]]))

plt.imshow(image)
plt.show()

plt.imshow(edges)
plt.show()
```

# Laplacian edge detection

```python
# Apply Gaussian blur to reduce noise
blurred = cv2.GaussianBlur(image, (3, 3), 0)

# Apply Laplacian edge detection
edges = cv2.Laplacian(blurred, cv2.CV_64F)

plt.imshow(image)
plt.show()

plt.imshow(edges)
plt.show()
```

# Canny edge detection

```python
# Apply Gaussian blur to reduce noise
blurred = cv2.GaussianBlur(image, (3, 3), 0)

# Apply Canny edge detection
edges = cv2.Canny(blurred, threshold1=100, threshold2=200)  # You can adjust the thresholds

plt.imshow(image)
plt.show()

plt.imshow(edges)
plt.show()
```
