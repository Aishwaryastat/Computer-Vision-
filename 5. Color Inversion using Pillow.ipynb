```python
import cv2
import numpy as np 
from  PIL import Image, ImageChops
import matplotlib.pyplot as plt
```

## Method 1:
Using inbuilt method ImageChops.invert() for Negating color.

```python
image= Image.open(r"C:\Users\pcc\Desktop\colorful.jpg", "r")

# image.show()

inv_image=ImageChops.invert(image)
plt.imshow(image)
plt.show()

plt.imshow(inv_image)
plt.show()
```

## Method 2:
The method used for getting the inverse of an image is subtraction of the maximum value/intensity of a pixel by the value of current pixel.

```python
img=np.array(image)

max_img=np.max(img)

inv_image_array = max_img-img

inv_image= Image.fromarray(inv_image_array)

plt.imshow(image)
plt.show()

plt.imshow(inv_image)
plt.show()
```
